```python
#txt-Datei einlesen

file = open('data.txt', 'r')
Lines = file.readlines()

measures = []
for line in Lines:
    measures.append(int(line))
```

```python
# Aufgabe 1
inc_counter = 0
for i in range(len(measures)):
    if i == 0:
        print(f"{measures[i]} (N/A - no previous measurement)")
    elif i > 0:
        if measures[i] > measures[i-1]:
            print(f"{measures[i]} increased")
            inc_counter += 1
        elif measures[i] < measures[i-1]:
            print(f"{measures[i]} decreased")
        elif measures[i] == measures[i-1]:
            print(f"{measures[i]} stayed same")

print(f"{inc_counter} increases counted")
```
