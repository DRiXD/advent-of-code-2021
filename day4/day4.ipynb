```python
#txt-Datei einlesen

import numpy as np

def read_bingo_txt(file_path):
    file = open(file_path, 'r')
    
    first_line = file.readline()
    stripped_line = first_line.strip()
    splitted_line = first_line.split(',')
    splitted_int_line = [int(i) for i in splitted_line]
    bingo_draws = splitted_int_line
    
    Lines = file.readlines()
    bingo_fields = []
    bingo_field = []
    for line in Lines:
        if(line!="\n"):
            stripped_line = line.strip()
            splitted_line = line.split()
            splitted_int_line = [int(i) for i in splitted_line]
            bingo_field.append(splitted_int_line)
        else:
            bingo_fields.append(np.array(bingo_field))
            bingo_field.clear()
    
    return bingo_draws, bingo_fields[1:]
```

```python
# Aufgabe 1

def cross_element_on_bingo_field(bingo_field, element):
    new_field = np.where(bingo_field == element, -1, bingo_field)
    return new_field

def check_if_bingo_field_wins(bingo_field):
        row_sums = bingo_field.sum(axis=1)
        col_sums = bingo_field.sum(axis=0)
        if(np.isin(-5, row_sums) or np.isin(-5, col_sums)):
            print("WINNER")
            print(bingo_field)
            return True
        else:
            return False

def play_bingo(bingo_draws, bingo_fields):
    for draw in bingo_draws:
        print("_____________")
        print(f"DRAW: {draw}")
        
        new_fields = []
        for bingo_field in bingo_fields:
            bingo_field = cross_element_on_bingo_field(bingo_field, draw)
            new_fields.append(bingo_field)
            
            #print(f"Bingo Field:")
            #print(bingo_field)
            #print("")
            
            if(check_if_bingo_field_wins(bingo_field)):
                break
        bingo_fields = new_fields
    return bingo_fields
        
        
bingo_draws, bingo_fields = read_bingo_txt("bingo.txt")
final_bingo_fields = play_bingo(bingo_draws, bingo_fields)
print(final_bingo_fields[0])
print("")
print(final_bingo_fields[1])
print("")
print(final_bingo_fields[2])
print("")
```
