```python
#txt-Datei einlesen

import numpy as np
import re

def read_vents_txt(file_path):
    file = open(file_path, 'r')
    Lines = file.readlines()
    
    coordinates = []
    for line in Lines:
        line = line.strip()
        coordinates.append(line)
    return coordinates
```

```python
# Aufgabe 1

coordinates = read_vents_txt("hydros.txt")

def calculate_points(coords):
    matches = re.findall(r'[0-9]+', coords)
    x_start = int(matches[0])
    x_end = int(matches[2])
    y_start = int(matches[1])
    y_end = int(matches[3])
    
    if x_start == x_end:
        return get_vertical_points(x_start, y_start, y_end)
    elif y_start == y_end:
        return get_horizontal_points(y_start, x_start, x_end)
    else:
        pass
    
        
def get_vertical_points(x, y_start, y_end):
    vertical_points = []
    for y in range(min(y_start,y_end), (max(y_start,y_end)+1)):
        point = (x,y)
        vertical_points.append(point)   
    return vertical_points


def get_horizontal_points(y, x_start, x_end):
    horizontal_points = []
    for x in range(min(x_start,x_end), (max(x_start,x_end)+1)):
        point = (x,y)
        horizontal_points.append(point)   
    return horizontal_points


points = []
for coords in coordinates:
    #print(f"Coords: {coords}")
    points.append(calculate_points(coords))
    points = [z for z in points if z is not None]
    #print(f"Points: {calculate_points(coords)}")
    #print("")

#print("----------------------------------")
vent_field = np.full((1000, 1000), 0, dtype=int)
for vent_line in points:
    for point in vent_line:
        vent_field[point[1],point[0]] += 1

#print(vent_field)

dangerous_areas = np.where( vent_field > 1 )
print(len(dangerous_areas[0]))
```

```python
# Aufgabe 2

coordinates = read_vents_txt("hydros.txt")

def calculate_points(coords):
    matches = re.findall(r'[0-9]+', coords)
    x_start = int(matches[0])
    x_end = int(matches[2])
    y_start = int(matches[1])
    y_end = int(matches[3])
    
    if x_start == x_end:
        return get_vertical_points(x_start, y_start, y_end)
    elif y_start == y_end:
        return get_horizontal_points(y_start, x_start, x_end)
    else:
        return get_diagonal_points(x_start, x_end, y_start, y_end)
    
        
def get_vertical_points(x, y_start, y_end):
    vertical_points = []
    for y in range(min(y_start,y_end), (max(y_start,y_end)+1)):
        point = (x,y)
        vertical_points.append(point)   
    return vertical_points


def get_horizontal_points(y, x_start, x_end):
    horizontal_points = []
    for x in range(min(x_start,x_end), (max(x_start,x_end)+1)):
        point = (x,y)
        horizontal_points.append(point)   
    return horizontal_points


def get_diagonal_points(x_start, x_end, y_start, y_end):
    diagonal_points = []
        
    if x_start > x_end and y_start > y_end:
        #print("a")
        while(x_start > x_end-1):
            while(y_start > y_end-1):
                point = (x_start,y_start)
                diagonal_points.append(point)
                x_start -= 1
                y_start -= 1
        #print(diagonal_points)
    elif x_start < x_end and y_start > y_end:
        #print("b")
        while(x_start < x_end+1):
            while(y_start > y_end-1):
                point = (x_start,y_start)
                diagonal_points.append(point)
                x_start += 1
                y_start -= 1
        #print(diagonal_points)
    elif x_start > x_end and y_start < y_end:
        #print("c")
        while(x_start > x_end-1):
            while(y_start < y_end+1):
                point = (x_start,y_start)
                diagonal_points.append(point)
                x_start -= 1
                y_start += 1
        #print(diagonal_points)
    elif x_start < x_end and y_start < y_end:
        #print("d")
        while(x_start < x_end+1):
            while(y_start < y_end+1):
                point = (x_start,y_start)
                diagonal_points.append(point)
                x_start += 1
                y_start += 1
        #print(diagonal_points)
    return diagonal_points


points = []
for coords in coordinates:
    #print(f"Coords: {coords}")
    points.append(calculate_points(coords))
    points = [z for z in points if z is not None]
    #print(f"Points: {calculate_points(coords)}")
    #print("")

#print("----------------------------------")
vent_field = np.full((1000, 1000), 0, dtype=int)
for vent_line in points:
    for point in vent_line:
        vent_field[point[1],point[0]] += 1

#print(vent_field)

dangerous_areas = np.where( vent_field > 1 )
print(len(dangerous_areas[0]))
```
